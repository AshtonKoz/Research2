```python
import numpy as np
import matplotlib.pyplot as plt
import math
import cmath
```

```python
h = 6.62607015e-34 #Planck's constant in Js
h_eV = h/1.6022e-19 #Planck's constant in eVs
hbar = h/(2*np.pi) #reduced Planck's constant
m = 9.1093837e-31 #electron mass in kg
```

```python
#Infinite Potenital Step
j = cmath.sqrt(-1)
s = 5000
E = 0.85e-38 #J
Vo = 0.75e-38 #J
Vo_array = np.linspace(0.75e-38,0.75e-38,s)
 
k = np.sqrt((2*m*E)/(hbar**2)) #electron wave number
print(k)
w = E/hbar #electron angular frequency
p = hbar*k #electron momentum
q = np.sqrt((2*m*(E-Vo))/(hbar**2)) #after potential wave number

x1 = np.linspace(-15,0,s)
x2 = np.linspace(0,15,s)

R = (k-q)/(k+q)
T = 2*k/(k+q)
print(R, T)
#assert 1 + R == T

psi_1_plane = np.cos(k*x1) + j*np.sin(k*x1)  
psi_1_reflec = R*(np.cos(k*x1) - j*np.sin(k*x1))
psi_2 = T*(np.cos(q*x2) + j*np.sin(q*x2))

plt.plot(x1,psi_1_plane, color='blue', label='Psi I')
plt.plot(x1,psi_1_reflec, color='green', label='Psi I reflected')
plt.plot(x2,psi_2, color='red', label='Psi II')
plt.plot(x2, Vo_array, color='black', linestyle = 'dashed')
plt.legend()
plt.xlabel('x')
plt.ylabel('Psi(x)')



#move V0 to other side, play with fluxes, plot flux, flux^2 vs energy
```
