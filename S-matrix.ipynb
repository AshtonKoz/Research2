```python
def truncate(f, n):
    '''Truncates/pads a float f to n decimal places without rounding'''
    s = '{}'.format(f)
    if 'e' in s or 'E' in s:
        return '{0:.{1}f}'.format(f, n)
    i, p, d = s.partition('.')
    return '.'.join([i, (d+'0'*n)[:n]])
```

## Fixed Incident energy, Varied potential depth

```python
E = 1.4e-29 #roughly 1uK
Vo = np.linspace(-1.4e-28,0,s) #potential well ~10000x as deep as incident energy
a = 1e-6 #one um
k = np.sqrt(2*m*E)*(1/hbar)
kappa = np.sqrt(2*m*(E-Vo))*(1/hbar)
delta = np.arctan((k/kappa)*np.tan(kappa*a)) - k*a
plt.plot(Vo, (delta/np.pi), color = 'red', label = f'$\delta$/$\pi$' )
plt.xlabel('Vo')
plt.ylabel(f'$\delta$/$\pi$')
plt.ylim(-0.6,0.6)
plt.legend()

print(max(delta/np.pi))
print(max(delta))
#print(((np.pi/2)-k*a)/np.pi)
```

```python


E1 = 1.4e-29 #roughly 1uK
Vo1 = np.linspace(-1.4e-10,0,s) #potential well 1000x as deep as incident energy
a = 1e-6 #one Angstrom
k1 = np.sqrt(2*m*E1)*(1/hbar)
kappa1 = np.sqrt(2*m*(E1-Vo1))*(1/hbar)
delta = np.arctan((k1/kappa1)*np.tan(kappa1*a)) - k1*a

sigma = ((4*np.pi)/k1**2)*((np.sin(delta))**2)

#plt.plot(Vo1, sigma, color = 'purple')
plt.plot(Vo1,(kappa*a)/(np.pi), color = 'red')
plt.xlabel('Vo')
plt.ylabel(f'$\sigma$')
plt.title(f'$\sigma$ vs Vo')

print(max(sigma))
```
