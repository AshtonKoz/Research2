```python
import numpy as np
import matplotlib.pyplot as plt
import math
import cmath
import scipy
from scipy import interpolate
s = 1000
```

# SI SECTION

```python
mrb = 1.4192261e-25 # rubidium mass in kg
h = 6.62607015e-34 # Planck's constant in Js
hbar = h/(2*np.pi) # Reduced Planck's constant in Js


E = np.linspace(1.380648780669e-32, 1.380648780669e-23,s) # particle energy from 1nK -> 1K in SI
Vo = -(10*1.380648780669e-23) # well depth of -10K in SI
a = 40*5.2918e-11 # 40 Bohr radii in m

k = np.sqrt(2*mrb*E/hbar**2) # wave number outside the well in m^-1
K = np.sqrt(2*mrb*(E-Vo)/(hbar**2)) # wave number inside the well in m^-1
d = np.arctan((k/K)*np.tan(K*a)) - k*a # phase shift
sigma = (np.pi/(k**2))*(np.sin(d))**2

Ko = np.sqrt(2*mrb*(-Vo))
a_s = a - (np.tan(Ko*a)/Ko)
sigma_approx = np.pi*(a_s)**2




#print(f' E = {E}')
#print(f' a_s = {a_s}ao')
#print(f' mass of rubidium = {mrb} au')
print(f' sigma approx = {sigma_approx/0.0001}')


plt.loglog(E/1.380648780669e-23, sigma/0.0001, label = f'$\\sigma$ SI', color = 'red')
plt.hlines(sigma_approx/0.0001, min(E), max(E), linestyle = 'dashed', color = 'black')
plt.legend()
plt.xlabel('particle energy (K)')
plt.ylabel(f'$\\sigma$ ($cm^2$)')
```

```python
mrb = 1.4192261e-25 # rubidium mass in kg
h = 6.62607015e-34 # Planck's constant in Js
hbar = h/(2*np.pi) # Reduced Planck's constant in Js
a = 40e-10 # 40A

Vo = -1e5*1.4e-29 # 1 million X ~1uK
E = np.linspace(1.4e-29, 0.95*abs(Vo),s) 

k = np.sqrt((2*mrb*E)/hbar**2)
K = np.sqrt((2*mrb*(E-Vo))/hbar**2)
Ko = np.sqrt((2*mrb*-Vo)/hbar**2)

delta = np.arctan((k/K)*np.tan(K*a)) - k*a
sigma = (np.pi/k**2)*((np.sin(delta))**2)

a_s = (a - (np.tan(a*Ko)/Ko))
delta_approx = -(a_s*k)
sigma_approx = np.pi*(a_s**2)

print(f' sigma approx = {sigma_approx/1e-28} barns')

print(f' a*Ko = {a*Ko}')
print(f' tan(a*Ko)/a*Ko = {np.tan(a*Ko)/(a*Ko)}')
print(f' tan is = {(np.tan(a*Ko)/Ko)}')
print(f' a_s = {a_s}')
print(f' min of k = {min(k)}, max of k = {max(k)}')

plt.hlines(sigma_approx*10000, min(E)/1.380648780669e-23, max(E)/1.380648780669e-23, color = 'black', linestyle = 'dashed', label = f"$\\sigma$ approx.")
SI2 = plt.loglog(E/1.380648780669e-23, sigma*10000, color = 'red', label = f"$\\sigma$ SI")
plt.legend()
plt.xlabel(f'particle energy (K)')
plt.ylabel(f'cross section ($cm^2$)')
```
